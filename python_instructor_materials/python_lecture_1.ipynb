## Jupyter Notebook

Jupyter Notebook is a **free, open-source, interactive web tool** known as a computational notebook, which researchers can use to combine code, computational output, text and multimedia resources in a single document. 

You can think of them as laboratory notebooks for scientific computing. The notebook can "speak" dozens of programming languages  reflected in its name inspired by the programming languages Julia (Ju), Python (Py) and R.

Analysis of the code-sharing site GitHub counted more than **2.5 million public Jupyter notebooks** in September 2018, up from 200,000 in 2015.

Example: [Gravitational Wave Open Science Center](https://github.com/losc-tutorial/quickview/blob/master/index.ipynb) ([Binder](https://hub.ovh2.mybinder.org/user/losc-tutorial-quickview-901j6od4/notebooks/index.ipynb))

## Jupyter notebooks are useful for...

* Really good for linear workflows (e.g. read data, filter data, do some statistics, plot the results)
* Sharing and explaining code to colleagues
* As an interactive development environment for code, data analysis, and visualization
* Prototyping: setting up an interactive environment to execute code, see what happens, modify and repeat
* Creating tutorials or interactive manuals for communicating data analysis routines
* Export as as a presentation (including this one)

## Pitfalls
* Non-linear code flow
* Large code-base (notebooks can be an entrypoint though)
* Limitations in version control (mostly resolved now)
